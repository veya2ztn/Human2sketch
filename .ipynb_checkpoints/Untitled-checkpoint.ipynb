```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
def gray_print(img):
    plt.imshow(img,'gray')
def rgb_print(img):
    img_rgb=img[:,:,[2,1,0]]
    plt.imshow(img_rgb,'gray')
import os
def under_file(file_dir):
    if not os.path.isfile(file_dir):
        filepaths=[]
        for filename in os.listdir(file_dir):
            if "csv" not in filename:
                filepath=os.path.join(file_dir,filename)
                filepaths.extend(under_file(filepath))
    else:
        filepaths=[file_dir]
    return filepaths
```

```python
cascade_file="/media/tianning/DATA/anime_data_sets/GirlsManifold-master/data/lbpcascade_animeface.xml"
```

```python
cascade = cv2.CascadeClassifier(cascade_file)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray = cv2.equalizeHist(gray)
faces = cascade.detectMultiScale(gray)
```

```python
img_h = image.shape[0]
img_w = image.shape[1]
```

```python
images = []
scale=1.3
min_size=30

image_rgb=image.copy()
for (x, y, w, h) in faces:
    scale_factor = min((scale - 1.) / 2, 
                       x / w, 
                       y / h,
                       (img_w - x - w) / w, 
                       (img_h - y - h) / h)
    scale_factor=scale-1
    sw = int(w * scale_factor *1)
    sh = int(h * scale_factor *1)
    ratio=10
    xx = x + w + sw;xx=min([xx,img_w])
    yy = y + h + sh/ratio;yy=min([yy,img_h])
    x = x - sw;x=max([x,0])
    y = y - sh*(ratio+9)/ratio;y=max([y,0])

    assert (x >= 0 and y >= 0 and xx <= img_w and yy <= img_h), (w, h, image.shape, x, y, xx, yy)
    image_rgb=cv2.rectangle(image_rgb, (x,y), (xx, yy), (7,249,151), 2)    
    if xx - x >= min_size and yy - y >= min_size:
        images.append(image[y:yy, x:xx, :])
rgb_print(image_rgb)
```
